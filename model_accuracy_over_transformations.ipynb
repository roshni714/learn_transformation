```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def make_figure_accuracy(xs, ys, name):
    name = name.capitalize()
    ylabel = "Model Accuracy"
    xlabel = "{} Parameter".format(name)
    title = "CIFAR-10: {} of Test Set vs. Model Accuracy".format(name)
    fig = plt.figure()
    plt.plot(xs, ys)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.savefig("figs/{}_vs_acc.pdf".format(name))
    plt.show()
```

```python
def make_figure_embed_dist(xs, ys, std, name):
    name = name.capitalize()
    ylabel = "Avg Distance to Training Set"
    xlabel = "{} Parameter".format(name)
    title = "CIFAR-10: {} of Test Set vs. Avg Distance to Training Set".format(name)
    fig = plt.figure()
    plt.plot(xs, ys)
    plt.errorbar(xs, ys, std)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.savefig("figs/{}_vs_embed_dist.pdf".format(name))
    plt.show()
```

```python
def make_figure_embed_dist_train(xs, ys, std, name):
    name = name.capitalize()
    ylabel = "Avg Distance to Training Set (from Train Set)"
    xlabel = "{} Parameter".format(name)
    title = "CIFAR-10: {} vs. Avg Distance to Training Set".format(name)
    fig = plt.figure()
    plt.plot(xs, ys)
    plt.errorbar(xs, ys, std)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.savefig("figs/{}_vs_embed_dist_train.pdf".format(name))
    plt.show()
```
